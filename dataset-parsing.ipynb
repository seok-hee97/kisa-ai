```python
import pandas as pd

# Load the dataset
df_full = pd.read_csv('A-Track-dataset.csv')

# Function to extract HTTP request components
def extract_http_components(payload):
    components = payload.split(' ')
    if len(components) == 3:
        method, path, protocol = components
        return pd.Series({'Method': method, 'Path': path, 'Protocol': protocol})
    else:
        return pd.Series({'Method': None, 'Path': None, 'Protocol': None})

# Apply the function to extract components and add as new columns
df_components = df_full['payload'].apply(extract_http_components)
df_full = pd.concat([df_full, df_components], axis=1)

# Display the updated DataFrame
print(df_full.head())
```

```python
import pandas as pd

# Load the dataset
df_full = pd.read_csv('A-Track-dataset.csv')

# Function to parse HTTP request components from payload
def parse_http_request(payload):
    # Split the payload by spaces to extract components
    components = payload.split(' ')
    
    # Extract method, path, and protocol
    method = components[0] if len(components) > 0 else None
    path = components[1] if len(components) > 1 else None
    protocol = components[2] if len(components) > 2 else None
    
    return method, path, protocol

# Apply the function to parse HTTP request components and add as new columns
df_full['Method'], df_full['Path'], df_full['Protocol'] = zip(*df_full['payload'].apply(parse_http_request))

# Display the updated DataFrame
print(df_full.head())
```

```python
import re

def parse_http_payload(payload):
  """HTTP 구성요소별로 payload를 분리합니다.

  Args:
    payload: HTTP payload

  Returns:
    분리된 HTTP 구성요소
  """

  # HTTP 구성요소의 시작과 끝을 식별하는 정규식
  patterns = {
      "method": r"^[A-Z]+\s",
      "url": r"^[A-Z]+\s\S+\s",
      "http_version": r"HTTP/\d+\.\d+",
      "headers": r"(?P<header>[\w-]+:\s[\w-]+\s)(?:(?P<value>[\w-]+\s)*)",
      "body": r"(?P<body>.*)"
  }

  # payload를 분리
  results = []
  for pattern in patterns.keys():
    match = re.search(pattern, payload)
    if match:
      results.append(match.groupdict()[pattern])

  return results

# 예시
payload = "GET /sub_04_1_read.php?page=1%27%20WHERE%204636%3D4636%20OR%209569%3D%28SELECT%20COUNT%28%2A%29%20FROM%20INFORMATION_SCHEMA.COLUMNS%20A%2C%20INFORMATION_SCHEMA.COLUMNS%20B%2C%20INFORMATION_SCHEMA.COLUMNS%20C%29--%20kIOu&id=31 HTTP/1.1\r\nCache-Control: no-cache\r\nUser-Agent: sqlmap/1.6.10#stable (https://sqlmap.org)\r\nReferer: http://10.0.17.20:80/sub_04_1_read.php\r\nHost: 10.0.17.20\r\nAccept: */*\r\nAccept-Encoding: gzip,deflate\r\nConnection: close\r\n\r\n"

parsed_payload = parse_http_payload(payload)

print(parsed_payload)
```

```python
import pandas as pd
import re

# Read the CSV file into a Pandas DataFrame
df_full = pd.read_csv('A-Track-dataset.csv')

# Define a regular expression to split the payload into the different components of an HTTP request
HTTP_REQUEST_REGEX = re.compile(r'^(?P<method>\w+)\s(?P<path>[-_\w./?:]+)\s(?P<version>HTTP/\d\.\d)\r\n(?P<headers>.*?)\r\n\r\n(?P<body>.*)', re.IGNORECASE)

# Extract the different components of the HTTP request
method = df_full['payload'].apply(lambda payload: HTTP_REQUEST_REGEX.search(payload).group('method'))
path = df_full['payload'].apply(lambda payload: HTTP_REQUEST_REGEX.search(payload).group('path'))
query_parameters = df_full['payload'].apply(lambda payload: HTTP_REQUEST_REGEX.search(payload).group('query_parameters'))
headers = df_full['payload'].apply(lambda payload: HTTP_REQUEST_REGEX.search(payload).group('headers'))
body = df_full['payload'].apply(lambda payload: HTTP_REQUEST_REGEX.search(payload).group('body'))

# Add new columns to df_full for each component of the HTTP request
df_full['method'] = method
df_full['path'] = path
df_full['query_parameters'] = query_parameters
df_full['headers'] = headers
df_full['body'] = body

# Print the head of the updated DataFrame
print(df_full.head())
```

```python
import pandas as pd
import re

def extract_http_components(request):
    components = {}
    
    # Extract HTTP method (GET or POST)
    method_match = re.match(r'^([A-Z]+)\s+', request)
    if method_match:
        components['method'] = method_match.group(1)
    
    # Extract Host
    host_match = re.search(r'Host:\s+([^\r\n]+)', request)
    if host_match:
        components['host'] = host_match.group(1)
    
    # Extract Path (assuming path is the part after the host in the request line)
    path_match = re.match(r'^[A-Z]+\s+[^/]+(/[^?\s]+)', request)
    if path_match:
        components['path'] = path_match.group(1)
    
    # Extract User Agent
    user_agent_match = re.search(r'User-Agent:\s+([^\r\n]+)', request)
    if user_agent_match:
        components['user_agent'] = user_agent_match.group(1)
    
    # Extract Cookie
    cookie_match = re.search(r'Cookie:\s+([^;\r\n]+)', request)
    if cookie_match:
        components['cookie'] = cookie_match.group(1)
    
    # Extract Content-Type
    content_type_match = re.search(r'Content-Type:\s+([^;\r\n]+)', request)
    if content_type_match:
        components['content_type'] = content_type_match.group(1)
    
    # Extract Referrer
    referer_match = re.search(r'Referer:\s+([^\r\n]+)', request)
    if referer_match:
        components['referer'] = referer_match.group(1)
    
    # Extract Accept
    accept_match = re.search(r'Accept:\s+([^;\r\n]+)', request)
    if accept_match:
        components['accept'] = accept_match.group(1)
    
    # Extract Accept-Language
    accept_language_match = re.search(r'Accept-Language:\s+([^;\r\n]+)', request)
    if accept_language_match:
        components['accept_language'] = accept_language_match.group(1)
    
    # Extract Accept-Encoding
    accept_encoding_match = re.search(r'Accept-Encoding:\s+([^;\r\n]+)', request)
    if accept_encoding_match:
        components['accept_encoding'] = accept_encoding_match.group(1)
    
    # Extract the remaining text as the body
    remaining_text_match = re.search(r'\r\n\r\n(.+)$', request, re.DOTALL)
    if remaining_text_match:
        components['body'] = remaining_text_match.group(1)
    
    return components

# Sample DataFrame
# data = {
#     'Log_Number': [1, 2, 3],
#     'payload': [
#         'POST /path1 HTTP/1.1\r\nContent-Type: application/json\r\n\r\n{"name": "John Doe"}',
#         'GET /path2 HTTP/1.1\r\nAccept: text/html\r\n\r\n',
#         'POST /path3 HTTP/1.1\r\nContent-Type: application/x-www-form-urlencoded\r\n\r\nkey1=value1&key2=value2'
#     ],
#     'label_action': ['action1', 'action2', 'action3']
# }

# df_full = pd.DataFrame(data)
df_full = pd.read_csv('A-Track-dataset.csv')
# Apply the extraction function and add the components to the DataFrame
df_full = df_full.join(df_full['payload'].apply(lambda x: pd.Series(extract_http_components(x))))

# Print the updated DataFrame
print(df_full)
```

```python
import pandas as pd
import re
import urllib.parse

def extract_http_components(request):
    components = {}
    
    # Extract HTTP method (GET or POST)
    method_match = re.match(r'^([A-Z]+)\s+', request)
    if method_match:
        components['method'] = method_match.group(1)
    
    # Extract Host
    host_match = re.search(r'Host:\s+([^\r\n]+)', request)
    if host_match:
        components['host'] = host_match.group(1)
    
    # Extract Path (assuming path is the part after the host in the request line)
    path_match = re.match(r'^[A-Z]+\s+[^/]+(/[^?\s]+)', request)
    if path_match:
        components['path'] = path_match.group(1)
    
    # Extract User Agent
    user_agent_match = re.search(r'User-Agent:\s+([^\r\n]+)', request)
    if user_agent_match:
        components['user_agent'] = user_agent_match.group(1)
    
    # Extract Cookie
    cookie_match = re.search(r'Cookie:\s+([^;\r\n]+)', request)
    if cookie_match:
        components['cookie'] = cookie_match.group(1)
    
    # Extract Content-Type
    content_type_match = re.search(r'Content-Type:\s+([^;\r\n]+)', request)
    if content_type_match:
        components['content_type'] = content_type_match.group(1)
    
    # Extract Referrer
    referer_match = re.search(r'Referer:\s+([^\r\n]+)', request)
    if referer_match:
        components['referer'] = referer_match.group(1)
    
    # Extract Accept
    accept_match = re.search(r'Accept:\s+([^;\r\n]+)', request)
    if accept_match:
        components['accept'] = accept_match.group(1)
    
    # Extract Accept-Language
    accept_language_match = re.search(r'Accept-Language:\s+([^;\r\n]+)', request)
    if accept_language_match:
        components['accept_language'] = accept_language_match.group(1)
    
    # Extract Accept-Encoding
    accept_encoding_match = re.search(r'Accept-Encoding:\s+([^;\r\n]+)', request)
    if accept_encoding_match:
        components['accept_encoding'] = accept_encoding_match.group(1)
    
    # Extract the remaining text as the body and decode URL
    remaining_text_match = re.search(r'\r\n\r\n(.+)$', request, re.DOTALL)
    if remaining_text_match:
        body = remaining_text_match.group(1)
        decoded_body = urllib.parse.unquote(body)
        components['body'] = decoded_body
    
    return components

# # Sample DataFrame
# data = {
#     'Log_Number': [1, 2, 3],
#     'payload': [
#         'POST%20/path1%20HTTP/1.1%0AContent-Type%3A%20application%2Fjson%0A%0A%7B%22name%22%3A%20%22John%20Doe%22%7D',
#         'GET%20%2Fpath2%20HTTP%2F1.1%0AAccept%3A%20text%2Fhtml%0A%0A',
#         'POST%20%2Fpath3%20HTTP%2F1.1%0AContent-Type%3A%20application%2Fx-www-form-urlencoded%0A%0Akey1%3Dvalue1%26key2%3Dvalue2'
#     ],
#     'label_action': ['action1', 'action2', 'action3']
# }

# df_full = pd.DataFrame(data)

# df_full = pd.DataFrame(data)
df_full = pd.read_csv('A-Track-dataset.csv')
# Decode URL in payload
df_full['payload'] = df_full['payload'].apply(urllib.parse.unquote)

# Apply the extraction function and add the components to the DataFrame
df_full = df_full.join(df_full['payload'].apply(lambda x: pd.Series(extract_http_components(x))))

# Print the updated DataFrame
print(df_full)
```
